```python
# workout_stats is a dictionary with string keys and tuple values - dictionary with tupples - keys and values
workout_stats = { 
    "Alex":(30, 45, 20),
    "Jamie":(40, 30, 25),
    "Taylor":(35, 50, 15)
}
```

```python
# calculate total workout miutes for each friend - using loops 
for friend, minutes in list(workout_stats.items()):
    if isinstance(minutes, tuple):
        total_minutes = sum(minutes)
        workout_stats[friend + "_Total"] = total_minutes
```

```python
#workot_list is a 2d nested list of workout minutes

workout_list = []

for minutes in workout_stats.values():
    if isinstance(minutes,  tuple):
        workout_list.append(list(minutes))
```

```python
#asking user to eter a friends name

friend_name = input("Enter a friend's name: ")

#check if friend exists 

if friend_name in workout_stats and isinstance(workout_stats[friend_name], tuple):
    yoga, running, weightlifting = workout_stats[friend_name]
    total = workout_stats[friend_name + "_Total"]

    print("Workout minutes:")
    print("Yoga:", yoga)
    print("Running:", running)
    print("Weightlifting:", weightlifting)
    print("Total minutes", total)

else:
    print(f"Friend {friend_name} not found in the records.") 

    
```

```python
#find person with highest and lowest total workout minutes

highest_friend = None
lowest_friend = None
highest_minutes = None
lowest_minutes = None

for key, value in workout_stats.items():
    if key.endswith("_Total"):
        if highest_minutes is None or value > highest_minutes:
            highest_minutes = value
            highest_friend = key.replace("_Total", "")

        if lowest_minutes is None or value < lowest_minutes:
            lowest_minutes = value
            lowest_friend = key.replace("_Total", "")


print("Friend with highest total workout minutes:" , highest_friend, "-", highest_minutes)
print("Friend with lowest total workout minutes:" , lowest_friend, "-", lowest_minutes)
```
